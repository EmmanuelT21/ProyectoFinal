```python
import random
import numpy as np
import matplotlib.pyplot as plt

def multiplos_fila(x):
    if x > 0 and x % 1 == 0:
        aleatorio = [random.randint(1, 9) for _ in range(x)]
        multiplos = [[i * num for i in range(1, x + 1)] for num in aleatorio]
        return aleatorio, multiplos
    else:
        return None, None

def multiplos_columna(x):
    if x > 0 and x % 1 == 0:
        aleatorio = [random.randint(1, 9) for _ in range(x)]
        multiplos = [[num * i for num in aleatorio] for i in range(1, x + 1)]
        return aleatorio, multiplos
    else:
        return None, None

def graficar_funciones():
    x = np.linspace(-2*np.pi, 2*np.pi, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    y3 = x

    plt.plot(x, y1, 'co-', label='sin(x)')
    plt.plot(x, y2, 'gd-', label='cos(x)')
    plt.plot(x, y3, 'bv-', label='x')

    plt.xlabel("Eje X")
    plt.ylabel("Eje Y")
    plt.title("Gráfica de funciones")
    plt.legend()
    plt.grid(True)
    plt.show()

N = str(5)
if N.isdigit():
    N = int(N)
    if N > 0:
        v_fila, resultado_fila = multiplos_fila(N)
        if resultado_fila:
            print("Vector fila:", v_fila)
            for fila in resultado_fila:
                print(fila)

        v_columna, resultado_columna = multiplos_columna(N)
        if resultado_columna:
            print("Vector columna:", v_columna)
            for columna in resultado_columna:
                print(columna)

        graficar_funciones()
    else:
        print("Error: Debe ingresar un número entero positivo.")

print("Error: Debe ingresar un número entero positivo.")
```
